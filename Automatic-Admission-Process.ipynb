# Introduced My Program ::-

A government wants to automate the admission process in all engineering colleges. A
government collected all relevant data and store it in CSV file. 

First File College_details.csv columns like CollegeCode,No of seat for each branch,Rank given by government.

Second file is studentdetails.csv which has information about the students result and their
enrolment key. columns like student enrolment key and 5 subject marks.

Third file has choice_filling.csv which has student admission choice details columns like enrolment
key and 10 preferable CollegeID-Branch code in which he/she likes to take admission.

1) First Generate the college_details.csv and studentdetails.csv file. Generate the atleast 100
   colleges and 10000 students data.
  
2) Give rank to the students so that admission process will be performed in proper sequence.

3) ask user to give atleast 3 preferable branch code in which they want to take admission and generate the data 
   based on the branch choice entered by student and store in the choice_filling.csv file.
  
4) First generate the data for 500 students and try to give them admission in their preferable college and branch.

5) Every day, the admission process of 500 students is carried out by the system.

6) At the end of first day of admission process, generate day1_college_status.csv file, which has details like number 
   of vacant seats, branch wise, in all the colleges that will available for admission process on day 2.

7) Run this System until entire admission process get over.

8) So, Finally Generate the report which shows the collegewise-branchwise first and last enrolment number which can be helpful
   to next year students. And Also publish rank to each branch which is highly preferred by student for current year admission.
